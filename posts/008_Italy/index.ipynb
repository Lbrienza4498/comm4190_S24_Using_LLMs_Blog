# My Dream Italian Getaway:

After my graduation in May, I have a celebratory trip planned with my two sisters to the motherland, Italy! However, the only things we have booked are our roundtrip flights from JFK to Milan. Since we are planning on doing a multi-city trip, incorporating Milan, Lake Camo, Portofino, and the Amalfi Coast, our heads are all spinning on how we are going to find the time to plan. So, I put my thinking cap on (kinda) and decided to ask AI for help. Similar to the example we were shown in class, where we used AI to replace a Starbucks barista, I want to use an AI tool and ask it to simulate a travel agent. The goal is to maximize our amount of time in each city while also finding lodging and recommendations that accommodate our interests in nightlife and proximity to the beach when applicable. 


## The Initial Plan

My tentative itinerary looked something like this:

1. Arrive in Milan on 6/10, spend a day and night exploring the city
2. Travel to Lake Como on 6/11 for a 3-night stay

4. Fly from Genoa to Naples and get to the Amalfi Coast on 6/16 or 6/17 for a 5-6 night stay
